```python
import pandas as pd
df=pd.read_csv('C:\TS_cntnt_example\dates_extraction.csv')
```

```python
from datetime import date
#using 'to_datetime' TO Transform string to date 
df['date_joined']=pd.to_datetime(df['date_joined'])
df
```

```python
#LETS EXTRACT DAY  FROM DATE and making it a new colum
df['DAY']=df['date_joined'].dt.day

#EXTRACT MONTH  FROM DATE and making it a new colum
df['MONTH']=df['date_joined'].dt.month

#EXTRACT YEAR  FROM DATE and making it a new colum
df['YEAR']=df['date_joined'].dt.year

#EXTRACT name of the day FROM DATE and making it a new colum
df['DAY NAME']=df['date_joined'].dt.day_name()

df
```
